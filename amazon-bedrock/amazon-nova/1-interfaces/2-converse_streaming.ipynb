```python
import boto3
import json

client = boto3.client('bedrock-runtime')

model_id = "amazon.nova-lite-v1:0"

messages = [
    {
        "role": "user",
        "content": [
            {
                "text": "じゃがいもと豚肉を使った晩御飯のおかずのレシピを3つ、それぞれ200文字以内で教えてください。"
            }
        ]
    }
]

system = [
    {
        "text": "あなたは家庭料理の得意な料理人です。"
    }
]

inferenceConfig = {
    "maxTokens": 1000,
    "temperature": 0.7,
    "topP": 0.9
}

response_stream = client.converse_stream(
    modelId=model_id,
    system=system,
    messages=messages,
    inferenceConfig=inferenceConfig
)

print("===== Raw Response =====")
print(response_stream)

print("=====  Final output =====")
for chunk in response_stream["stream"]:
    if "contentBlockDelta" in chunk:
        text = chunk["contentBlockDelta"]["delta"]["text"]
        print(text, end="")
```
