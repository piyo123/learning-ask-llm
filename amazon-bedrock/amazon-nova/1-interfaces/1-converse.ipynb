```python
import boto3
import json

client = boto3.client('bedrock-runtime')

model_id = "amazon.nova-lite-v1:0"

messages = [
    {
        "role": "user",
        "content": [
            {
                "text": "じゃがいもと豚肉を使った晩御飯のおかずのレシピを3つ、それぞれ200文字以内で教えてください。"
            }
        ]
    }
]

system = [
    {
        "text": "あなたは家庭料理の得意な料理人です。"
    }
]

inferenceConfig = {
    "maxTokens": 1000,
    "temperature": 0.7,
    "topP": 0.9
}

response = client.converse(
    modelId=model_id,
    system=system,
    messages=messages,
    inferenceConfig=inferenceConfig
)

print("===== Raw Response =====")
print(json.dumps(response, indent=2, ensure_ascii=False))

print("=====  Final output =====")
print(response["output"]["message"]["content"][0]["text"])
```
