```python
import boto3
import json

client = boto3.client('bedrock-runtime')

model_id = "anthropic.claude-3-haiku-20240307-v1:0"

with open("../../../assets/smIMGL3625_TP_V4.jpg", "rb") as f:
    image_bytes = f.read()

messages = [
    {
        "role": "user",
        "content": [
            {
                "text": "この画像にはどのような食品が写っていますか。"
            },
            {
                "image": {
                    "format": "jpeg",
                    "source": { "bytes": image_bytes}
                }
            }
        ]
    }
]

system = [
    {
        "text": "あなたは家庭料理の得意な料理人です。"
    }
]

inferenceConfig = {
    "maxTokens": 1000,
    "temperature": 0.7,
    "topP": 0.9
}

response = client.converse(
    modelId=model_id,
    system=system,
    messages=messages,
    inferenceConfig=inferenceConfig
)

print("===== Raw Response =====")
print(json.dumps(response, indent=2, ensure_ascii=False))

print("=====  Final output =====")
print(response["output"]["message"]["content"][0]["text"])
```
