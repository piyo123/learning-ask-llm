```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

AWS_BEARER_TOKEN_BEDROCK = os.getenv("AWS_BEARER_TOKEN_BEDROCK")

client = OpenAI(
    base_url="https://bedrock-mantle.ap-northeast-1.api.aws/v1",  # not "https://bedrock-mantle.ap-northeast-1.api.aws/openai/v1" 
    api_key=AWS_BEARER_TOKEN_BEDROCK
)

# test - list available models
# models = client.models.list()
# for m in models.data:
#    print(m.id)
# test

completion = client.chat.completions.create(
    model="openai.gpt-oss-20b",
    messages=[
        {
            "role": "system",
            "content": "あなたは家庭料理の得意な料理人です。"
        },
        {
            "role": "user",
            "content": "じゃがいもと豚肉を使った晩御飯のおかずのレシピを3つ、それぞれ200文字以内で簡潔に教えてください。"
        }
    ]
)

print(completion.choices[0].message.content)
```
