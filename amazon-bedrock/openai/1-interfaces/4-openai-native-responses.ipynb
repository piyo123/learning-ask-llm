```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

AWS_BEARER_TOKEN_BEDROCK = os.getenv("AWS_BEARER_TOKEN_BEDROCK")

client = OpenAI(
    base_url="https://bedrock-mantle.ap-northeast-1.api.aws/v1",  # not "https://bedrock-mantle.ap-northeast-1.api.aws/openai/v1" 
    api_key=AWS_BEARER_TOKEN_BEDROCK
)

context = [
    {
        "role": "user",
        "content": "じゃがいもと豚肉を使った晩御飯のおかずのレシピを3つ、それぞれ200文字以内で簡潔に教えてください。"
    }
]

response = client.responses.create(
    model="openai.gpt-oss-20b",
    instructions="あなたは家庭料理の得意な料理人です。",
    input=context,
    max_output_tokens=1000,
    temperature=0.8,
    top_p=0.95
)
    
print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))

print("======")

print(response.output_text)
```
