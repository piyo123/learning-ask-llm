```python
import os
import json
from openai import AzureOpenAI
from dotenv import load_dotenv

load_dotenv()
AZURE_OPENAI_ENDPOINT = os.getenv("AZURE_OPENAI_ENDPOINT") # ex. https://project1.openai.azure.com/
AZURE_OPENAI_API_KEY = os.getenv("AZURE_OPENAI_API_KEY")
AZURE_OPENAI_API_VERSION = "2025-03-01-preview"
MODEL_DEPLOYMENT_NAME = "gpt-4o-mini"

client = AzureOpenAI(
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_key=AZURE_OPENAI_API_KEY,
    api_version=AZURE_OPENAI_API_VERSION
)

context = [
    {
        "role": "user",
        "content": "滋賀県について教えてください。"
    }
] 

response = client.responses.create(
    model=MODEL_DEPLOYMENT_NAME,
    instructions=("あなたは親しみやすい観光ガイドです。\n"
                         "提示された観光地の特徴や魅力をそれぞれ、文化、歴史、自然、エンターテインメントの観点で説明してください。"
                ),
    input=context,
    max_output_tokens=1000,
    temperature=0.8,
    top_p=0.95
)
    
print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
print()
print(response.output_text)

# ref: https://platform.openai.com/docs/guides/migrate-to-responses?update-multiturn=responses#3-update-multi-turn-conversations
context += response.output

context += [
    { "role": "user", "content": "そうなんですね、東京からどれぐらいの時間がかかりますか？"}
]

response2 = client.responses.create(
    model=MODEL_DEPLOYMENT_NAME,
    input=context,
    max_output_tokens=500,
)

print()
print(json.dumps(
        response2.model_dump(),
        indent=2,
        ensure_ascii=False
    )
)

print()
print(response2.output_text)
```
