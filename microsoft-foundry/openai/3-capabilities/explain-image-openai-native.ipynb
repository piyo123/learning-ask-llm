```python
import os
import json
import base64
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
OPENAI_ENDPOINT = os.getenv("OPENAI_ENDPOINT") # ex. https://project1.openai.azure.com/openai/v1/
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
MODEL_DEPLOYMENT_NAME = "gpt-4o-mini"

client = OpenAI(
    base_url=OPENAI_ENDPOINT,
    api_key=OPENAI_API_KEY
)

with open("../../../assets/smIMGL3625_TP_V4.jpg", "rb") as f:
    img_b64 = base64.b64encode(f.read()).decode("utf-8")

data_url = f"data:image/jpg;base64,{img_b64}"

context = [
        {
            "role": "user",
            "content": [
                { "type": "input_text", "text": "この画像にはどのような食品が写っていますか。"},
                { 
                    "type": "input_image",
                    "image_url": data_url
                }
            ]
        }
] 

response = client.responses.create(
    model=MODEL_DEPLOYMENT_NAME,
    instructions="あなたは親切で思慮深いアシスタントです。",
    input=context,
    max_output_tokens=1000,
    temperature=0.8,
    top_p=0.95
)
    
print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
print("----------------------------------------------------------------")
print(response.output_text)
```
